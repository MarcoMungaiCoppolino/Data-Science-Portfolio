```python
import googleapiclient.discovery
import os

# Set API key environment variable
os.environ["GOOGLE_API_KEY"] = "AIzaSyDnySFPWSEqsh5BHJT9w9BomsmMv5s0kEE"

# Create YouTube API client
youtube = googleapiclient.discovery.build("youtube", "v3", developerKey=os.environ["GOOGLE_API_KEY"])

# Search for videos related to AI and Chat GPT
request = youtube.search().list(
        q="AI, ChatGPT",
        type="video",
        part="id,snippet",
        maxResults=1,
        order="viewCount"
)
response = request.execute()

# Get channel, subscriber count, view count, and comment count for each video
for video in response["items"]:
    video_id = video["id"]["videoId"]
    video_info = youtube.videos().list(
            part="snippet,statistics",
            id=video_id
    ).execute()

    channel_name = video_info["items"][0]["snippet"]["channelTitle"]
    subscriber_count = video_info["items"][0]["statistics"].get("subscriberCount", 0)
    view_count = video_info["items"][0]["statistics"]["viewCount"]
    comment_count = video_info["items"][0]["statistics"]["commentCount"]
    print(video_info["items"][0]["snippet"] )
    print(f"Video: {video_info['items'][0]['snippet']['title']}")
    print(f"Channel: {channel_name}")
    print(f"Subscriber count: {subscriber_count}")
    print(f"View count: {view_count}")
    print(f"Comment count: {comment_count}")
    print()
```
