```python
for i in range(0, len(GLT)):
    j_prec = i-1
    j_ant = i+1
    
    if GLT[i][1] == '':
        while j_prec != -2 or j_ant != -2: 
            if j_prec != -2:
                if j_prec == -1: 
                    # if there are no avaible elements before the while will continue to subtract -1 until reaching -1
                    # in that case we set prec = 0
                    prec = 0
                    j_prec = -2
                    break
                elif GLT[j_prec][1] != '' and GLT[j_prec][3] == GLT[i][3]:
                    prec = float(GLT[j_prec][1])
                    j_prec = -2 
                else:
                    j_prec -= 1
            
            if j_ant != -2:
                if j_ant == len(GLT):
                    # same as before but this time if there are no elements j will increase until reaching the end of the list
                    ant = 0
                    j_ant = -2
                    break
                elif GLT[j_ant][1] != '' and GLT[j_ant][3] == GLT[i][3]:
                    ant = float(GLT[j_ant][1])
                    j_ant = -2
                else: 
                    j_ant += 1                   
        GLT[i][1] = str((prec + ant)/2)
```

```python
count = 0
for i in range(0, len(GLT)):
    if GLT[i][1] == '':
        count += 1

print(count)
```
